```python
import tensorflow as tf

from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt
```

```python
import h5py
import numpy as np
```

```python
data = h5py.File('h5/images2.h5', 'r')
```

```python
images = data['images'][:]
target = data['target'][:]
```

```python
target = np.where(target==1)[1]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(images, target)
```

```python
x_train = x_train[:,:,:, np.newaxis]/255.0
x_test = x_test[:,:,:, np.newaxis]/255.0
```

```python
y_train
```

```python
model = models.Sequential()
model.add(layers.Conv2D(70, (3, 3), activation='relu', input_shape=(70, 70, 1)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(140, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(140, (3, 3), activation='relu'))
```

```python
model.add(layers.Flatten())
model.add(layers.Dense(140, activation='relu'))
model.add(layers.Dense(8))
```

```python
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])

history = model.fit(x_train, y_train, epochs=10, 
                    validation_data=(x_test, y_test))
```

```python
import cv2 as cv
import numpy as np
```

```python
classes = model.predict(x_test, batch_size=5)
preds = np.argmax(classes, axis=1)
```

```python
classes_name = ['anger', 'contempt', 'disgust', 'fear', 'happiness', 'neutral', 'sadness', 'surprise']
```

```python
for i in range(5):
    plt.imshow(x_test[i][:,:,0])
    predict = "\nKt qu d on: " + str(classes_name[preds[i]])
    output = "\nlabel:"+ str(classes_name[y_test[i]])
    plt.text(0, -2, predict)
    plt.text(0, -4, output)
    plt.show()
```

```python
# save model as json
model_json = model.to_json()
with open("model.json", "w") as json_file:
    json_file.write(model_json)

model.save_weights("model.h5")
print("Saved model to disk")
```
